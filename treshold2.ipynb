## Калькуляция угла между конечностями

```python
def calculate_angle(a,b,c):
    a=np.array(a)
    b=np.array(b)
    c=np.array(c)

    radians=np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    angle=np.abs(radians*180.0/np.pi)

    if angle>180.0:
        angle=360-angle
    return angle
```

## Подсчет повторений левой руки


```python
#  Video Capturing
cap = cv2.VideoCapture(0)
counter =0
stage =None


with mp_pose.Pose(min_detection_confidence =0.5,min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        # изменение цветов картинки 
        image= cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image.flags.writeable=False

        # Детекция
        results=pose.process(image)
        # покраска картинки обратно в RGB
        image.flags.writeable=True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)

        # Извлечение ключевых точек
        try:
            landmarks= results.pose_landmarks.landmark
            ## Координаты 
            shoulder_left=[landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow_left = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist_left = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            angle_left=calculate_angle(shoulder_left,elbow_left, wrist_left)
            #Визуализация
            cv2.putText(image,str(angle_left),
                        tuple(np.multiply(elbow_left,[640,480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255),2,cv2.LINE_AA
                        )
            
            # Counter 
            if angle_left>160:
                stage='down'
            if angle_left<30 and stage =='down':
                stage='up'
                counter+=1
                print(counter)
        except:
            pass
        
        
        cv2.rectangle(image,(0,0),(240,73),(245,117,16),-1)
# Повторения
        cv2.putText(image,'Reps_L',(15,12),
                    cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1,cv2.LINE_AA)
        cv2.putText(image,str(counter),
                   (5,60),
                   cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),2,cv2.LINE_AA           )
       
       
        #  В какой стадии
        cv2.putText(image,'Stage',(110,12),
                    cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1,cv2.LINE_AA)
        cv2.putText(image,stage,
                   (80,60),
                   cv2.FONT_HERSHEY_SIMPLEX, 2,(255,255,255),2,cv2.LINE_AA           )
        #  В какой стадии

        ## Рендеринг детекции
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                  )


        cv2.imshow('Mediapipe Feed', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
